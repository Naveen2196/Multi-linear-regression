# Problem Statement1(50_Startups.csv)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d5 = pd.read_csv("50_Startups.csv")
d5
```

```python
d5.info()
```

# LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
d5['State'] = le.fit_transform(d5['State'])
```

# Boxplot

```python
d5.plot.box()
```

# Pairplot

```python
sns.pairplot(d5)
```

# Droping Target variable

```python
#For better prediction don't transform taget variable or dependent feature into standardization
dd5 = d5[['Profit']]
d5 = d5.drop(columns = 'Profit',axis = 1)
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
scaled_data = ss.fit_transform(d5)
sd5 = pd.DataFrame(scaled_data,columns = ['RDSpend','Administration','MarketingSpend','State'])
```

# Ordinary least_square method

```python
sns.histplot(sd5,kde = True,bins = 15)
```

# Adding 'Profit' column to Existing Dataset

```python
sd5['Profit'] = dd5['Profit']
sd5
```

```python
import statsmodels.formula.api as smf
model = smf.ols('Profit~RDSpend+MarketingSpend+State',data = sd5).fit()
model.summary()
```

# Importing Libraries

```python
# Import necessary libraries
from sklearn.model_selection   import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
```

# Train_Test_Split

```python
#extract feature and target column data
features_data = sd5[['RDSpend','Administration','MarketingSpend','State']]
target_data = sd5[['Profit']]
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features_data, target_data, test_size=0.2, random_state=42)
```

# Regularization of model

```python
# Initialize the Ridge Regressor with alpha value (regularization strength)
ridge = Ridge(alpha=0.1)

# Fit the model on your data
ridge.fit(X_train, y_train)

# Predict on unseen data
#predictions = ridge.predict(X_test)
```

# Logistic regression

```python
# Create a Linear Regression model
lre = LinearRegression()

# Train the model
lre.fit(X_train, y_train)

# Use the model to predict the test set
y_pred = lre.predict(X_test)

# Print the predicted values
print(f'prediction values : {y_pred}')

# Compute the accuracy of the actual model
act_accry = lre.score(X_train, y_train)

# Print the accuracy score
print(f'\nActual Accuracy: {act_accry*100}')


# Compute the accuracy of the model
accuracy = lre.score(X_test, y_test)

# Print the accuracy score
print(f'Prediction Accuracy: {accuracy*100}\n')

# Compute the Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Print the RMSE score
print(f'Root Mean Squared Error: {rmse}\n')

# Compute normalized RMSE
normalized_rmse = rmse / (np.max(y_test) - np.min(y_test))

print(f'Normalized RMSE: {normalized_rmse}')

#This produces a value between 0 and 1,
#where values closer to 0 represent better fitting models
```

# Calculating VIF

```python
rsq_hp = smf.ols('RDSpend~Administration+MarketingSpend+State+Profit',data=sd5).fit().rsquared
vif_hp = 1/(1-rsq_hp)

rsq_wt = smf.ols('Administration~MarketingSpend+State+Profit+RDSpend',data=sd5).fit().rsquared  
vif_wt = 1/(1-rsq_wt) 

rsq_vol = smf.ols('MarketingSpend~State+Profit+RDSpend+Administration',data=sd5).fit().rsquared  
vif_vol = 1/(1-rsq_vol) 

rsq_sp = smf.ols('State~Profit+RDSpend+Administration+MarketingSpend',data=sd5).fit().rsquared  
vif_sp = 1/(1-rsq_sp) 

rsq_profit = smf.ols('Profit~RDSpend+Administration+MarketingSpend+State',data=sd5).fit().rsquared  
vif_profit = 1/(1-rsq_profit) 

# Storing vif values in a data frame
dd5 = {'Variables':['RDSpend','Administration','MarketingSpend','State','Profit'],'VIF':[vif_hp,vif_wt,vif_vol,vif_sp,vif_profit]}
Vif_frame = pd.DataFrame(dd5)  
Vif_frame
```

# Test for Normality of Residuals (Q-Q Plot)

```python
import statsmodels.api as sm

model = smf.ols('Profit~RDSpend+Administration+MarketingSpend+State',data=sd5).fit()
qqplot=sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

# Residual Plot for Homoscedasticity

```python
model = smf.ols('Profit~RDSpend+Administration+MarketingSpend+State',data=sd5).fit()
```

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

# Residual Vs Regressors

# Detecting Influencers/Outliers

# Cook’s Distance

```python
from statsmodels.graphics.regressionplots import influence_plot

model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
c
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(sd5)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Identify points with high Cook's Distance
high_cook_dist = np.where(c > np.mean(c) + 3 * np.std(c))
```

```python
# Print the indices of the influential points
print("Influential points:", high_cook_dist)
```
